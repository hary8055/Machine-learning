```python
train,test=train_test_split(df,test_size=0.1)
```

```python
trainx=train.iloc[:,:8]
trainy=train.iloc[:,8]
testx=test.iloc[:,:8]
testy=test.iloc[:,8]

# model building
lg=LogisticRegression()
predlg=lg.fit(trainx,trainy).predict(testx)

# accuracy of model
accuracy=np.mean(predlg==testy)
accuracy
```

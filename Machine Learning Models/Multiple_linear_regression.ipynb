```python
#libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('/Users/harikrishnans/Downloads/StudentsPerformance.csv')
```

```python
df
```

```python
# check for null value
df.isnull().sum()
```

```python
# check for dupplicates
df.duplicated().sum()
```

```python
# check for dupplicates
df.duplicated().sum()
```

```python
# removing dummy variable
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dummy=['gender','race/ethnicity','parental level of education','lunch','test preparation course']
for i in dummy:
    df[i]=le.fit_transform(df[i].astype(str))
df
```

```python
x=df.iloc[:,[0,1,2,3,4,6,7]]
x
```

```python
y=df.iloc[:,5]
y
```

```python
# splitting to train and test
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=True)

# model building
lr=LinearRegression()
lr.fit(xtrain,ytrain)
predlr=lr.predict(xtest)
predlr
```

```python
# accuracy of model
score=r2_score(predlr,ytest)
score
```

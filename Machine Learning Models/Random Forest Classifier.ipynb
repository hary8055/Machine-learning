```python
#libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier as DTC
from sklearn.ensemble import RandomForestClassifier as RFC
```

```python
df = pd.read_csv('/Users/harikrishnans/Downloads/classification.csv')
```

```python
# check for null value
df.isnull().sum()
```

```python
# check for duplicates
df.duplicated().sum()
```

```python
# splitting to train and test
train,test=train_test_split(df,test_size=0.2)
trainx=train.iloc[:,0:2]
trainy=train.iloc[:,2]
testx=test.iloc[:,0:2]
testy=test.iloc[:,2]

# model building
dt=DTC(criterion='entropy')
modeldt=dt.fit(trainx,trainy).predict(testx)
modeldt
```

```python
# accuracy of decision tree
acc=np.mean(modeldt==testy)
acc
```

```python
# plotting tree
from sklearn import tree
tree.plot_tree(dt)
```

```python
# Random forest model building
rf=RFC(n_jobs=2,n_estimators=100,oob_score=True,criterion='entropy')
predrf=rf.fit(trainx,trainy).predict(testx)
predrf
```

```python
# random forest accuracy
acc1=np.mean(predrf==testy)
acc1
```
